#### 4.7.4 Removing Outliers <a class = "anchor" id = "Outliers" ></a>

Removing outliers is also a part of data preparation. Outliers are data points that skew the results of the data analysis, since outliers have either very large or very small value. To determine which data points are outliers, we will use some statistical concepts, such as quartiles, and interquartile range. Quartiles are the data points that divides the dataset into four equal parts. First quartile is the data point that divides the top 75% of the data to the bottom 25% of the data, while the third quartile divides the top 25% of the data to the bottom 75% of the data. Interquartile range is the difference between the third and first quartile.
Of all the columns in our data frame, only the column `duration` fits the bill of having outliers or not, so we will analyze `duration`. To get the first and third quartile, we will use the function `quantile()`, using `probs = 0.25` for the first quartile and `probs = 0.75` for the third quartile. We will use the function `iqr()` to get the interquartile range.

##### R Script:

```r
first_quartile <- quantile(trip_data_clean$duration, probs = 0.25)
third_quartile <- quantile(trip_data_clean$duration, probs = 0.75)
iqr <- IQR(trip_data_clean$duration)
```

In order for us to determine the data points that are considered outliers, we will use IQR method, which means that the data points are considered outliers if they are greater than the third quartile plus 1.5 times the interquartile range, or less than the first quartile minus 1.5 times the interquartile range.
$$
\text{Outlier} > Q_3 + 1.5IQR\\ \\ \text{OR}\\ \\
\text{Outlier} < Q_1 - 1.5IQR
$$

Let's compute the boundaries. I will assign names for the boundaries to make it easier for the next procedure.

##### R Script:

```r
boundary_1 <- as.double(round(first_quartile,2)) - 1.5*iqr
boundary_2 <- as.double(round(third_quartile,2)) + 1.5*iqr
```

Now that we know the boundaries in order to determine the outliers, we will remove the values that is less than `boundary_1` or greater than `boundary_2`.

##### R Script:

```r
trip_data_clean <- trip_data_clean %>% filter(duration > boundary_1 & duration < boundary_2)
```

#### 4.7.5 Removing Data with N/A <a class = "anchor" id = "NA" ></a>

The last step that I will do to clean the dataset is to remove the data points where there is a missing value or having N/A as its value on any of its column. I will use the function `na.omit()`.

##### R Script:

```r
trip_data_clean <- na.omit(trip_data_clean)
```

There are several insights that will surface if we analyze the duration of the ride, such as the average duration of the ride, the average duration of the ride per day of the week, the average duration of the ride for each type of bike, etc. Take note that the duration of the ride is in minutes.

To better understand the dataset, we can analyze first the average duration of the ride, and summary statistics are the best parameters to solve for. The summary statistics that I will calculate are:

* **Mean** - the arithmetic average duration of the ride. The R function for this is `mean()`.
* **Median** - the middle value of the duration of the ride, if the dataset is sorted from smallest to largest, or vice versa. The R function for this is `median()`.
* **Mode** - the duration with the greatest number of observations. 
* **Minimum** - the lowest duration of the ride. The R function for this is `min()`.
* **Maximum** - the highest duration of the ride. The R function for this is `max()`.

All of the summary statistics stated above have a function in R, except for the mode. Therefore, I wil create a code that can be used to solve for the mode.

##### R Script:

```r
mode <- function(x) {
  uniq_x <- unique(x)
  uniq_x[which.max(tabulate(match(x, uniq_x)))]
}
```

Now that we have a function for each of the summary statistics stated above, I will use these functions, as well as the function `summarise()` to tabulate all the summary statistics.
##### R Script:

```r
trip_data_clean %>%
    summarise(mean = mean(duration),
    median = median(duration),
    mode = mode(duration),
    min = min(duration),
    max = max(duration))
```


Now that we are done getting the summary statistics for the duration of the ride, we can now create some visualizations for better understanding of the distribution of the duration of the ride. One visualization that can help us in determining the distribution is using a histogram. I will use the function `hist()` to plot a histogram.
##### R Script:


```r
options(repr.plot.width = 10, repr.plot.height = 10)
histogram <- hist(trip_data_clean$duration,main = "Distribution of Duration of Ride", xlim = c(0,40),ylim = c(0,600000), xlab = "Duration (mins)", ylab = "Frequency", col = "#4095A5")
options(scipen=999)
```

### 5.7 Comparison of the Number of Rides per Month Between Casual Riders and Members <a class = "anchor" id = "Analyze7" ></a>

In order for us to create meaningful insights to solve the company's problem, we need to determine the month with the largest number of rides, and compare the numbers for casual riders and members. I will use again the function `ggplot()` to create a line chart. But first, I will create a temporary column for the month when the ride started. In creating a temporary column, I will use the function `mutate()`.
##### R Script:

```r
options(repr.plot.width = 15, repr.plot.height = 10)
trip_data_clean %>%
    mutate(month = month(start_time)) %>%
    group_by(membership,month) %>%
    summarise(number_of_rides = n(), .groups = "drop") %>%
    ggplot(aes(month, number_of_rides, color = membership)) +
    geom_line() +
    geom_point() +
    scale_color_manual(values = c("#4095A5","#A54095"))+
    scale_y_continuous(labels = scales::label_number_si(),
                        breaks = seq(0, 360000, 20000)) +
    scale_x_continuous(labels = c("Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sept","Oct","Nov","Dec"),breaks = seq(1,12,1))+
    labs(title = "Number of Rides per Month", x = "Month", y = "Number of Rides")
```

We can see that riders tend to start biking from March and peaks at the month of July and August, then the trend is decreasing until December of the year. I think this is the case since after September, the weather starts to grow cold, which is not the perfect time for biking. The trend is the same for both casual riders and members.

```r
options(repr.plot.width = 15, repr.plot.height = 10)
trip_data_clean %>%
    mutate(day_of_week = wday(start_time,label=TRUE)) %>%
    group_by(membership,day_of_week,hour_started) %>%
    summarise(number_of_rides = n(), .groups = "drop") %>%
    ggplot(aes(hour_started, number_of_rides, color = membership)) +
    geom_line() +
    geom_point() +
    scale_color_manual(values = c("#4095A5","#A54095"))+
    scale_y_continuous(labels = scales::label_number_si(),
                        breaks = seq(0, 360000, 10000)) +
    scale_x_continuous(breaks = seq(0,23,2))+
    labs(title = "Number of Rides per Day", x = "Hour", y = "Number of Rides")+
    facet_wrap(~day_of_week,ncol = 4)
    
```
